```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,6*np.pi,0.05)
rand = np.random.rand(377)
```

```python
f = np.sin(x)+rand/5
g = np.cos(x)+rand/5
```

```python
plt.plot(x,f,'b',label='sin(t) + noise')
plt.plot(x,g,'r',label='cos(t) + noise')
plt.xlabel('time t',fontsize=14)
plt.ylabel('Amplitude f(t)',fontsize=14)
plt.title('Curves with noise',fontsize=14)
plt.legend(loc='upper left')
```
