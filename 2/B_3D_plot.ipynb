```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
x=np.arange(-3,3,0.05)
```

```python
y=np.arange(-3,3,0.05)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
z =(1 * (2 * np.pi) ** (-0.5) * np.exp((X ** 2 + Y ** 2)/ (-2) ))
```

```python
fig=plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X,Y,z,rstride=1,cstride=10,cmap=cm.coolwarm)
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.xlabel('x axis',fontsize=14,color='k')
plt.ylabel('y axis',fontsize=14,color='k')
plt.title('gauss distrivution',fontsize =14,color='k')
```
